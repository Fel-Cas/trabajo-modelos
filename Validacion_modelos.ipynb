```python
## Importamos las librerias

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from joblib import dump, load
from sklearn.metrics import precision_score,recall_score,accuracy_score,f1_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df=pd.read_csv("/DatosReales.csv")
df
```

```python
## Separamos los datos
X=df.iloc[:,1:]
Y=df.iloc[:,0]
X["BMI"] = encoder.fit_transform(X["BMI"])
```

```python
def validarModelo(model, X, Y):
  y_pred=model.predict(X)
  print("Accuracy:",metrics.accuracy_score(Y, y_pred))
  print(f'Precission Score: {precision_score(Y,y_pred)}')
  print(f'Recall Score: {recall_score(Y,y_pred)}')
  print(f'F1-Score: {f1_score(Y,y_pred)}')
  cm = confusion_matrix(Y, y_pred, labels=model.classes_)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
  disp.plot()
```

# Gradient Boosting

```python
model = load('/content/GradientBoosting.joblib')
validarModelo(model,X,Y)
```

# Gradient Boosting con OverSampling

```python
model = load('/content/GradientBoosting_OverSampling.joblib')
validarModelo(model,X,Y)
```

```python
model = load('/content/GradientBoosting_UnderSampling.joblib')
validarModelo(model,X,Y)
```
