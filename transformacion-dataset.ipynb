# Importamos el dataset

```python
df=pd.read_csv('/kaggle/input/dataset/heart_2020_cleaned.csv')
df.head()
```

## Eliminamos la columna de raza
Decidimos eliminar la columna que nos indica la raza de la persona, porque pensamos que este dato no debería influir si una persona tiene probabilidades de sufrir un ataque cardiaco o no, ya que muchos estudios han demostrado que esto depende más del estilo de vida de la persona.

```python
df=df.drop(["Race"],axis=1)
df.head(5)
```

# Transformación de los datos categóricos a numéricos

Empezamos con la transformación de los datos que tienen como respuesta Yes o No.   Estos datos los vamos a transformar a valores de 0 y 1.  
Las respuestas con valores de Yes, van a ser reemplazados por 1.  
Las respuestas con valores de No, va a ser reemplazados por 0.  

```python
df_categorico=df.select_dtypes(include=["object"])
```

```python
df_categorico.columns
```

```python
columns=['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking','PhysicalActivity',
       'GenHealth', 'Asthma', 'KidneyDisease', 'SkinCancer']
for column in columns:
    df[column].loc[df[column]=="Yes"]=1
    df[column].loc[df[column]=="No"]=0
```

La columna que nos indica si la persona ha tenido diabetes tiene más categorías como: No, borderline diabetes y Yes (during pregnancy).   
Como esta varaible nos indica que si la persona ha sufrido de diabetes a lo largo de su vida, decidimos transformar la categoría Yes (during pregnancy) por 1 y la categoría No, borderline diabetes por 0.

```python
categoria_diabetes=["Yes (during pregnancy)","Yes","No, borderline diabetes","No"]
for categoria in categoria_diabetes:
    if "No" in categoria:
        df["Diabetic"].loc[df["Diabetic"]==categoria]=0
    else:
        df["Diabetic"].loc[df["Diabetic"]==categoria]=1
df.head()
```

En la  columna GenHealth tenemos las siguientes categorias Very good, Fair, Good, Poor y Excellent. Nosotros vamos a transformar esas categorías a datos numericos de la siguiente manera:       
* Poor va a ser reemplazado por 2
* Fair va a ser reemplazado por 3
* Good va a ser reemplazado por 4
* Very good va a ser reemplazado por 5
* Excellent va a ser reemplazado por 6


```python
df["GenHealth"].unique()
```

```python
estado_salud=['Poor','Fair','Good','Very good','Excellent']
tipo_salud=2
for estado in estado_salud:
    df["GenHealth"].loc[df["GenHealth"]==estado]=tipo_salud
    tipo_salud+=1
df.head()
```

En la columna de Sex tenemos Female y Male.  
* Female va a ser reemplazado por 8
* Male va a ser reemplazado por 9

```python
sexos=df["Sex"].unique()
auxiliar=8
for sexo in sexos:
    df["Sex"].loc[df["Sex"]==sexo]=auxiliar
    auxiliar+=1
```

* "70-74"  va a ser reemplazado por 20

```python
df["AgeCategory"].unique()
```

```python
categoria_edad=["18-24","25-29","30-34","35-39","40-44","45-49","50-54","55-59","60-64","65-69","70-74", "75-79", "80 or older"]
auxiliar=10
for edad in categoria_edad:
    df["AgeCategory"].loc[df["AgeCategory"]==edad]=auxiliar
    auxiliar+=1
df.head()
```
